
# Definition of Factorial
The factorial of a non-negative integer n is defined as the product of all positive integers less than or equal to 
n
n. It is represented by the symbol 
n
!
n!.
### The factorial of a non-negative integer \( n \) is calculated using the formula n!=n×(n−1)×(n−2)×…×3×2×1. For example 5!=5×4×3×2×1=120 and by convention,  0! = 1.

```python
n= 5 

f =1  
for i in range (1, n+1):     
    f = f*i
print (f)    # factorial calculation
```

# Definition of a Function in Python
### A function in Python is a named block of code designed to perform a specific task. Functions help organize code, making it more modular and reusable.

## How to Define a Function
To define a function, you use the following syntax:

```python
def factorial (n): # here we named factorial function as factorial but other arbitary names are also allowed.
    f=1
    for i in range (1,n+1):
        f=f*i
    return f     # return is used to assign any computed outcome of the function into f value expected and passed down for further use   
    
```

```python
factorial (4), factorial (5), factorial (1), factorial (2), factorial (3), factorial (0) # samples of calling the funct. factorial
```

```python
f9 = factorial (9)
f0 = factorial (0)

print (f9,f0)     # here we assigned the values of the factorial 9 and zero to new variables f9 & f0 respectively and printed them both
```

## To convert polar coordinates to Cartesian coordinates in Python, you can use the following function: in numpy library
### To convert polar coordinates (r,ϕ) to Cartesian coordinates (x,y) in Python, you can use the following function:

```python
import numpy as np
def polar_to_cartesian(r, phi):
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return x, y
```

```python
r = 5
phi = np.pi / 4  # 45 degrees
x, y = polar_to_cartesian(r, phi)
print (x,y)
```

## For converting Cartesian coordinates to polar coordinates, use this function
### To convert Cartesian coordinates (x,y) back to polar coordinates (r,ϕ), use this function:

arctan2 (often written atan2) is a two-argument arctangent function that returns the angle (in radians) whose tangent is y/x, while using the signs of both x and y to determine the correct quadrant.

Key points:

    * Signature (common in many languages): atan2(y, x)

    * Returns angle θ in range (-π, π] (implementation-dependent; some use [0, 2π)).

    * Handles x = 0 safely (unlike atan(y/x)).

    * Useful for converting Cartesian (x, y) → polar (r, θ): θ = atan2(y, x), r = sqrt(x2 + y2).

```python
def cartesian_to_polar(x, y):
    r = np.sqrt(x**2 + y**2)  # ** In Python, ** is the exponentiation operator (power).Examples: 2 ** 4 == 16

    phi = np.arctan2(y, x)  #radians
    return r, phi
```

```python
x = 3
y = 4
r, phi = cartesian_to_polar(x, y)
print (r,phi)
```

## 1. Basic 2D plot (mathematical function)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 400)
y = np.sin(x) / x
plt.plot(x, y, label=r'$\mathrm{sinc}(x)=\sin x/x$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.axhline(0, color='k', lw=0.5)
plt.show()
```

## 2. Plotting vector fields (physics: electric/magnetic fields, flow)

```python
    
    # python  retains the last call values for import numpy as np & import matplotlib.pyplot as plt above


X, Y = np.meshgrid(np.linspace(-2,2,20), np.linspace(-2,2,20))
Ex = X / (X**2 + Y**2 + 0.1)   # example field components
Ey = Y / (X**2 + Y**2 + 0.1)
plt.streamplot(X, Y, Ex, Ey, density=1.2, arrowsize=1)
plt.xlabel('x'); plt.ylabel('y'); plt.axis('equal'); plt.show()
```

These create equally spaced numeric arrays and a 2D grid:

Example uses:
- x: plotting a 1D function y=f(x).
- t: time samples for time series / ODE solutions.
- X,Y: evaluate and plot surfaces or vector fields, e.g. Z = f(X,Y), plt.contourf(X,Y,Z) or plt.streamplot(X,Y,Ex,Ey).

## 3. Solving ODEs and plotting (damped harmonic oscillator)

```python

from scipy.integrate import solve_ivp

def dh(t, y):
    x, v = y
    k, m, c = 2.0, 1.0, 0.3
    return [v, -(k/m)*x - (c/m)*v]

t = np.linspace(0, 20, 1000)
sol = solve_ivp(dh, (t[0], t[-1]), [1.0, 0.0], t_eval=t)
plt.plot(sol.t, sol.y[0])
plt.xlabel('t'); plt.ylabel('x(t)'); plt.grid(True); plt.show()
```

##  5. Plot: electric field wave (sinusoidal) with trigonometric time signature

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1.0          # amplitude (V/m)
f = 5.0          # frequency (Hz)
omega = 2 * np.pi * f
phi = np.pi/6    # phase shift
t = np.linspace(0, 1, 1000)  # 1 second

# Electric field (sinusoidal wave)
E = A * np.sin(omega * t + phi)

# Trigonometric time signature example: envelope + modulation
envelope = 0.8 * (1 + 0.5 * np.cos(2 * np.pi * 1.0 * t))  # slow cosine envelope (1 Hz)
modulated = envelope * np.sin(omega * t + phi)           # amplitude modulation

plt.figure(figsize=(8,4))
plt.plot(t, E, label='Pure sinusoid: $E(t)=A\\sin(\\omega t+\\phi)$', alpha=0.7)
plt.plot(t, modulated, label='AM signal: envelope$\\cdot\\sin(...)$', color='C1')
plt.plot(t, envelope, '--', color='C2', label='Envelope (positive)')
plt.plot(t, -envelope, '--', color='C2', label='Envelope (negative)')
plt.xlabel('Time (s)')
plt.ylabel('Electric field (arb. units)')
plt.title('Electric Field Wave — Sinusoid with Trigonometric Time Signature')
plt.legend()
plt.grid(True)
plt.xlim(0, 0.2)   # zoom into first 0.2 s for clarity
plt.tight_layout()
plt.show()
```


Notes:
- Adjust A, f, envelope frequency, and phi to model different signals.
- For complex waves, sum multiple sinusoids or add phase shifts.

## Concise runnable examples showing LaTeX-style math in labels and annotations (uses r'...' raw strings) plus plt.annotate and plt.text.

### 1. Simple function with LaTeX label and inline text:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,400)
y = np.sin(x)/x
plt.plot(x, y)
plt.xlabel(r'$x$ (units)')
plt.ylabel(r'$y=\mathrm{sinc}(x)=\frac{\sin x}{x}$')
plt.title(r'Plot of $\mathrm{sinc}(x)$')
plt.text(1.5, 0.6, r'Peak near $x\approx 0$', fontsize=10)
plt.grid(True)
plt.show()
```

### 2. Annotate a point with an arrow and LaTeX expression:

```python


t = np.linspace(0,4*np.pi,400)
y = np.sin(t)
plt.plot(t, y)
peak_t = np.pi/2
peak_y = 1.0
plt.annotate(r'$\sin\!\left(\frac{\pi}{2}\right)=1$',
             xy=(peak_t, peak_y), xytext=(peak_t+2, 0.8),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=11)

plt.xlabel('t (s)')
plt.ylabel(r'$\sin t$')
plt.title('Sine wave with annotation')
plt.show()
```

### 3. 2D field with contour, LaTeX label and annotation of a saddle/peak:

```python


X, Y = np.meshgrid(np.linspace(-3,3,200), np.linspace(-3,3,200))
Z = np.sin(np.sqrt(X**2+Y**2)) / (np.sqrt(X**2+Y**2)+1e-9)
fig, ax = plt.subplots()
cs = ax.contourf(X, Y, Z, levels=30, cmap='viridis')
cbar = fig.colorbar(cs, ax=ax)
cbar.set_label(r'$Z(x,y)$ (arb. units)')
ax.set_title(r'Radial wave: $Z=\frac{\sin(\sqrt{x^2+y^2})}{\sqrt{x^2+y^2}}$')
# annotate a ring maximum
ax.annotate(r'Local max', xy=(1.5,0), xytext=(2.0,1.0),
            arrowprops=dict(arrowstyle='->', color='white'), color='white')
plt.show()
```

###  4. Annotating vector field with LaTeX and inline text:

```python
x = np.arange (0,20,0.1)
y = np.cos(x)
y1 = np.exp(-x/10)*np.cos(x)

plt.plot(x,y, label = "Oscillation i", linewidth= 3)
plt.plot(x,y1, label = "Oscillation ii", linewidth = 3,color = "r", linestyle= "--" )
plt.xlabel("$x$  axis", size =16)
plt.ylabel("$\\ cos(x)$", size =16)  # plt.ylabel(r'$\cos(x)$', size=16)

plt.legend ()
plt.show ()                      #example from Ernazar Abdikamalov
```

##### Two cosine waves vs. x from 0 to 20. 

- the blue curve (Oscillation i) is a pure cosine with constant amplitude; 
- the red dashed curve (Oscillation ii) is the same cosine multiplied by an exponential decay envelope exp(−x/10), so its amplitude decreases with x. 
- both share the same frequency and phase; 
- the legend, axis labels (x and cos(x)), and line styles distinguish them.

```python
x= np.arange(1,20,0.1)
y= (10**x)*(np.cos(x)**2)
plt.plot(x,y)

plt.show()
```

# Description of the plotted math function:

- y(x) = 10^x · cos^2(x) is a rapidly growing oscillatory function. 
- The cos^2(x) factor oscillates between 0 and 1 with period π, modulating the exponential envelope 10^x 
  which increases super-exponentially on a linear x scale. 
- As x increases the peaks (at cos^2=1) grow exponentially, troughs stay at 0, and values quickly overflow standard floating-point ranges. 
- On a log (y) scale the plot appears as a series of equally spaced oscillatory peaks riding on a straight line with slope log10(e)·
- ln(10) (i.e., linear growth in log10(y) proportional to x)
