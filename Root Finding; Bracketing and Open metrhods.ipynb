```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Equation:  $3x^3  + x-3 = 0$

```python
def f(x):    
    return 3.0*x**3+x-3 
```

```python
x =np.arange(0.5,1.51,0.1)
plt.plot(x,f(x))
plt.grid()  # グリッドを表示  gridline() is a function in Matplotlib that adds a grid to the plot, making it easier to read and interpret the data points. By calling plt.grid(), you enable the grid lines on the plot, which can help in visualizing the relationship between the x and y values more clearly.
plt.axhline(0, color='black', lw=0.5)   # x軸を引く  plt.axhline(y=0) draws a horizontal line at y=0, which is the standard x-axis
plt.show()
```

```python
def bisect (xl,xu,es,imax):
    iter = 0
    xr = xl
    while iter < imax:
        xrold = xr
        xr = (xl + xu) / 2.0
        iter += 1
        if xr != 0:
            ea = abs((xr - xrold) / xr) * 100.0
        test = f(xl) * f(xr)
        if test < 0:
            xu = xr
        elif test > 0:
            xl = xr
        else:
            ea = 0.0
        if ea < es:
            break
    return xr, iter, ea
```
